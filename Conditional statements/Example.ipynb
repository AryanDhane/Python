## If, Else, Elif Statements

```python
age = 20

if age >= 18:
    print("You are eligible to vote.")
elif age < 0:
    print("You are children and not eligible to vote.")    
else:
    print("You are not eligible to vote.")
```

## Nestad conditional statements

```python
# you can place one or more if, elif or else statements inside another if, elif or else statement. This is called nested if statement.

num = 10

if num > 0:
    print("Positive number")
    if num % 2 == 0:
        print("Even number")
    else:
        print("Odd number")
elif num == 0:
    print("Zero")
else:
    print("Negative number")

```

```python
#  calculator using if, elif, else statements

num1 = int(input("Enter num1: "))
num2 = int(input("Enter num2: "))

operator = input("Enter operator (+, -, *, /): ")
if operator == '+':
    print("Addition: ", num1 + num2)
elif operator == '-':
    print("Subtraction: ", num1 - num2)
elif operator == '*':
    print("Multiplication: ", num1 * num2)
elif operator == '/':
    print("Division: ", num1 / num2)
else:
    print("Invalid operator")
```
