### Converting one data type into another

 Python provides built-in functions:

    int()

    float()

    str()

    bool()

```python
x = "10"
y = int(x)
print(y + 5)  # 15
```

```python
int("abc")  # ❌ ValueError
```

```python
# INTEGER TO FLOAT
x = 10
y = float(x)
print(y)
```

```python
# Number to String
x = 10
y = str(x)
print(y)
```
