### input() always returns string

```python
a = input()
print(type(a))  # str
```

### Input with Type Casting

```python
age = int(input("Enter age: "))
salary = float(input("Enter salary: "))
print("Age:", age)
print("Salary:", salary)
```

### Multiple Inputs

```python
a, b = map(int, input().split())
print("a:", a)
print("b:", b)
```

```python
print(a, b, sep=",")
```

```python
print("Hello", end=" ")
print("World")
```
