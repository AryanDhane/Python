```python
# Declaring and Assigning Variables in Python

age = 19
height = 5.9
name = "Aryan"
is_student = True

print("Age:", age)
print("Height:", height)
print("Name:", name)
print("Is Student:", is_student)
```

```python
# Naming Conventations in Python
# Variable names should start with a letter or underscore. They can contain letters, digits, and underscores. Variable names are case-sensitive.

first_name = "Aryan"
last_name = "Dhane  "
print(first_name)
print(last_name)




```

```python
# Understanding variable types
# python is a dynamically typed, type of a variable is determined at runtime based on the value assigned to it.
 
age = 19
height = 5.9
name = "Aryan"
is_student = True
print(type(name))
print(type(height))
print(type(age))
```

```python
# type conversition
# converting one data type to another 

age = 19
print(type(age))

age_str = str(age)
print(type(age_str))
print(age_str)
```

```python
# Dynamic Typing
# python allowed the type of a variable to change during execution.
a = 13
print(a,type(a))
```
