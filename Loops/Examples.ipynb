## For loops

```python
# for loop
# the for loop will iterate over a sequence (like a list, tuple, string) or range

for i in range(11):
    print(i)
```

```python
for i in range(1, 6):
    print(i)
```

```python
for i in range(1, 11, 2): # start, stop, step
    print(i)
```

```python
#  strings

str1 = "Hello"
 
for i in str1:
    print(i)
```

## While Loops

```python
# while loops
# the loop will continue until the condition is false

count = 1
while count < 5:
    print(count)
    count += 1


```

```python
count = 0

while count % 2 == 0:
    print(count)
    count += 1
```

```python
#  Calculate the sum of first n numbers with while loop

n = 5
sum = 0
i = 1

while i <= n:
    sum += i
    i += 1

print(f"The sum of first {n} numbers is: {sum}")
```

## Loop Control Statements

#### ------ Break ------

```python
#  break statement

for i in range(10):
    if i == 5:
        break
    print(i)

```

#### ------ Continous ------

```python
#  continous statement
#  the continue statement skips the current iteration and moves to the next one

for i in range(10):
    if i == 5:
        continue
    print(i)
```

```python
# Pass Statement
# the pass statement is a null operation; 

for i in range(5):
    if i == 3:
        pass

    print(i)
```

## Nested Loop

```python
#  Nested Loops

for i in  range(4):
    for j in range(3):
        print(f"i: {i}, j: {j}")
        
```
