# The Map() Function in python

###  the map() function applies a given function to all items in an input list (or any iterable) and returns a map object (which is an iterator). this is particulariy useful for transforming data in a comprehensively.


```python
number = [1, 2, 3, 4, 5]

list(map(lambda x: x ** 2, number))
```

```python
names = ['aryan', 'rahul', 'sachin']
list(map(str.upper, names))
```
