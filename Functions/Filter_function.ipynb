```python
def even(num):
    if num%2 == 0:
        return True
```

```python
even(6)
```

```python
number=[1,2,3,4,5,6,7,8]
list(filter(even, number))
```

```python
number=[1,2,3,4,5,6,7,8]
greater_than_two = list(filter(lambda x:x>2, number))
print(greater_than_two)
```

```python
number=[1,2,3,4,5,6,7,8]
greater_than_two = list(filter(lambda x:x>2 and x%2==0, number))
print(greater_than_two)
```
